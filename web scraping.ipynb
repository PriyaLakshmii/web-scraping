```python
import requests
```

```python
data=requests.get("https://en.wikipedia.org/wiki/List_of_countries_by_coffee_production")
data.text
```

```python
from bs4 import BeautifulSoup as bs
data_bs=bs(data.text)
```

```python
table_data=data_bs.find('table')
```

```python
countries=[]
for row in table_data.find_all('tr')[1:12]:
    temp=row.text.replace('\n\n'," ").strip()
    temp=temp.split()
    countries.append(temp)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(countries,columns=['rank','countries','60 kg bags','metric tons','pounds'])
df
```

```python
df.to_excel("Coffee.xlsx")
```
